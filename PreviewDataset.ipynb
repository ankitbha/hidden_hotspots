### Comparing basic stats of gov vs our data

```python
use_seg = SEGMENTS[0]
(datamat, _), (_, _, names) = create_dataset_joint(
    use_seg, 
    split=1.0,
    exclude=[
        'Sirifort, New Delhi - CPCB',
        'Dr. Karni Singh Shooting Range, Delhi - DPCC',
        'Sri Aurobindo Marg, Delhi - DPCC',
    ])
```

```python
def plot_stats_vs(datamat, names, dsplit, height=4):
    dataA = datamat[:dsplit]
    dataB = datamat[dsplit:]
    
    plt.figure(figsize=(14, height*2))
    
    plt.subplot(2, 1, 1)
    mu = np.mean(dataA, axis=0)
    
    plt.plot(mu)
    
    
    mu = np.mean(dataB, axis=0)
    plt.plot(mu)
    
    plt.subplot(2, 1, 2)
    var = np.var(dataA, axis=0)
    plt.plot(var)
    var = np.var(dataB, axis=0)
    plt.plot(var)
    plt.ylim(0, 1)

    plt.show()
    plt.close()
    
plot_stats_vs(datamat, names, 9)
```
