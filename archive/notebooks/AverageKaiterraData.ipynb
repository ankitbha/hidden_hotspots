```python
# script to average the data to every 5 min/10 min/15 min interval as desired
# the logic is borrowed from the .do file in Dropbox that Sameeksha and Sujoy wrote
# called Kaiterra_2018_12_11_import.do

import pandas as pd
import pytz
INTERVAL = '15'
```

```python
inpath = '/home/shivar/Dropbox/Delhi Pollution/06_Data_Collection/Pilot_2018_Kaiterra_FieldEggs/'
fname = 'kaiterra_fieldeggs_all_1year.csv'
data = pd.read_csv(inpath + fname, index_col=[2,0], parse_dates=True)
```

```python
data
```

```python
data.rename(str.upper, axis=0, level=0, inplace=True)
data.index.levels[0], data.index.levels[1]
```

```python
# sort the dataframe by ids first, time next
data.sort_index(level=0, inplace=True)
data
```

```python
# set the timezone in data to be UTC so that the df becomes tz-aware
data.tz_localize(pytz.UTC, level=1, copy=False)
data
```

```python
# now, shift the timestamps in the data to the new timezone 
# (can be done only in tz-aware dataframe, hence previous step was necessary)
data.tz_convert('Asia/Kolkata', level=1, copy=False)
data
```

```python
# merge with locations in field_egg_details.dta
fieldeggsdetails = pd.read_stata('/home/shivar/Dropbox/Delhi Pollution/06_Data_Collection/Pilot_2018_Kaiterra_FieldEggs/Dta files/field_egg_details.dta')
fieldeggsdetails.udid = [s.replace('-','') for s in fieldeggsdetails.udid.values]
fieldeggsdetails.set_index('udid', inplace=True)
fieldeggsdetails
```

```python
data_merged = data.join(fieldeggsdetails[['latitude', 'longitude', 'comment']], on='device_udid')
data_merged.sort_index(level=0)
data_merged
```

```python
# for the averaging operation, first append a column with rounded timestamps
data_merged.reset_index(inplace=True)
data_merged['timestamp_round'] = data_merged.time.apply(lambda ts: ts.floor(INTERVAL + 'T'))
```

```python
# reindexing so that all timestamps are in same range for every sensor
start_ts = min(data_merged.timestamp_round)
end_ts = max(data_merged.timestamp_round)
print('Start ts:', start_ts)
print('End ts:', end_ts)
newindex = pd.date_range(start_ts, end_ts, freq=INTERVAL + 'T', tz='Asia/Kolkata', name='timestamp_round')

grouped = data_merged.groupby('short_id')
groups = []
for name, group in grouped:
    df = group.groupby('timestamp_round')[['pm_25','pm_10']].mean()
    df = df.reindex(newindex, axis=0)
    df['field_egg_id'] = name
    df['location'] = group.comment.iloc[0]
    df['device_udid'] = group.device_udid.iloc[0]
    df['latitude'] = group.latitude.iloc[0]
    df['longitude'] = group.longitude.iloc[0]
    df.reset_index(inplace=True)
    df.set_index(['field_egg_id', 'timestamp_round'], inplace=True)
    groups.append(df)

data_merged_avged = pd.concat(groups)
data_merged_avged.rename({'pm_25':'pm25', 'pm_10':'pm10'}, axis=1, inplace=True)
data_merged_avged
```

```python
data_merged_avged[['latitude', 'longitude', 'location', 'pm25', 'pm10']].to_csv('data/kaiterra_fieldeggid_{}min_2019_Feb_28_panel.csv'.format(INTERVAL))
```
