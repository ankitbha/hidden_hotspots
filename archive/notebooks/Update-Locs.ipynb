```python
# create locs.csv from the fieldeggdetails dta file
import pandas as pd
```

```python
fieldeggsdetails = pd.read_stata('/home/shivar/Dropbox/Delhi Pollution/06_Data_Collection/Pilot_2018_Kaiterra_FieldEggs/Dta files/field_egg_details.dta')
fieldeggsdetails.udid = [s.replace('-','') for s in fieldeggsdetails.udid.values]
fieldeggsdetails.set_index('udid', inplace=True)
fieldeggsdetails
```

```python
fieldeggsdetails['udid_short'] = [s[-4:].upper() for s in fieldeggsdetails.index.values]
fieldeggsdetails.udid_short
```

```python
fieldeggsdetails.rename({'comment':'location'}, axis=1, inplace=True)
fieldeggsdetails
```

```python
fieldeggsdetails = fieldeggsdetails.reset_index().set_index('udid_short')
fieldeggsdetails[['udid', 'latitude', 'longitude', 'location']].to_csv('../data/kaiterra/locs_new.csv')
```
