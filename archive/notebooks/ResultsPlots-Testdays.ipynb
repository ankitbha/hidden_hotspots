```python
import os
import glob
import numpy as np
import pandas as pd
```

```python
source = 'kaiterra'
sensor = 'pm25'
version = 'v1'
suffixdict = {'kaiterra':'seed9', 'govdata':'seed6'}
df = pd.read_csv('results/errors_{}_{}_{}_testdays_{}_alldays_0.txt'.format(source, sensor, version, suffixdict[source]), index_col=[0], parse_dates=True)
```

```python
df
```

```python
print(df.mape_mean.mean(), df.mape_mean.max(), df.mape_mean.min())
print(df.rmse_mean.mean(), df.rmse_mean.max(), df.rmse_mean.min())
```

```python
print(df.mape_min.mean(), df.mape_min.min(), df.mape_min.max())
print(df.rmse_min.mean(), df.rmse_min.min(), df.rmse_min.max())
```

```python
print(df.mape_max.mean(), df.mape_max.min(), df.mape_max.max())
print(df.rmse_max.mean(), df.rmse_max.min(), df.rmse_max.max())
```
