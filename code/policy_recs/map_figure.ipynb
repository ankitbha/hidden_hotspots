```python
import numpy as np
from PIL import Image
```

```python
map_fig = Image.open('./map_figure_policy.png')
```

```python
map_fig
```

```python
map_arr = np.array(map_fig)
```

```python
map_arr.shape
```

```python
map_long_lims = [76.745,77.796]
map_lat_lims = [28.261,28.845]
```

```python
crop_map = Image.fromarray(map_arr[150:650,250:1020,:])
```

```python
crop_map
```

```python
lat_range = map_lat_lims[1]-map_lat_lims[0]
crop_map_lat_lims = [map_lat_lims[0]+lat_range*(151/832),map_lat_lims[0]+lat_range*(650/832)]
long_range = map_long_lims[1]-map_long_lims[0]
crop_map_long_lims = [map_long_lims[0]+long_range*(251/1307),map_long_lims[0]+long_range*(1020/1307)]
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(8,12), dpi=100)
ax.imshow(np.array(crop_map),extent=[crop_map_long_lims[0],crop_map_long_lims[1],crop_map_lat_lims[0],crop_map_lat_lims[1]])
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
fig.savefig('lmap_policy.png')
```
