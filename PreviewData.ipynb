```python
scale = 0.4 # (x - min) / 0.4 = normalized under 1

ymin = df['latitude'].min()
xmin = df['longitude'].min()

lcount = {}
lkey = lambda xx, yy: '%.2f %.2f' % (xx, yy)
```

```python
for ind, (lat, lon) in df[['latitude', 'longitude']].iterrows():
#     print(ymin, xmin, lat, lon)
    ynorm = (lat - ymin) / scale
    xnorm = (lon - xmin) / scale
    kk = lkey(xnorm, ynorm)
#     print(kk)
    if kk not in lcount: lcount[kk] = 0
    lcount[kk] += 1
    
#     break

# print(len(lcount))
    
```

```python
lcount
```

```python
xs = []
ys = []
hs = []
sizes = []
zeros = []
for key, val in lcount.items():
    hs.append(val)
    xx, yy = key.split(' ')
    xs.append(float(xx))
    ys.append(float(yy))
    
    sizes.append(0.01)
    zeros.append(0)
    
fig = plt.figure(figsize=(14, 8))
ax1 = fig.add_subplot(111, projection='3d')
ax1.bar3d(xs, ys, zeros, sizes, sizes, hs, shade=True)
# ax1.set_title('Shaded')

ax1.view_init(45, 45)
plt.draw()

plt.show()
plt.close()
```

```python
timesegs = {}
lastend = {}
laststart = {}
# locsegs = {}
# maxskip = 6 * 60 * 60
maxskip = 6 * 60 # 6 minutes

for ind, (location, locname, tt, pm) in df[['loc_s_id', 'location', 'timestamp', 'pm25']].iterrows():
    location = '%s : %s' % (location, locname)
    if type(tt) == type(pd.NaT):
        continue
    dobj = pd.to_datetime(tt)
    seconds = time.mktime(dobj.timetuple())
#     location = lkey(lon, lat)
    if location not in timesegs: timesegs[location] = []

    if location not in lastend:
        # new appearance of this location
        lastend[location] = seconds
        laststart[location] = seconds
        continue
    else:
        diff = seconds - lastend[location]
        if diff > maxskip:
            # there was a huge skip, so end segment
            timesegs[location].append([laststart[location], lastend[location]])
            del lastend[location]
            del laststart[location]
        else:
            # data continues, extend segment
            lastend[location] = seconds
            pass
    
    if ind % 10 == 0:
        sys.stdout.write('%d/%d   \r' % (ind, -1))
        sys.stdout.flush()
        
#     if ind > 10000: break
        
for loc, _ in timesegs.items():
    # end any segmenst tracking through end of series
    if loc in lastend:
        assert loc in laststart
        timesegs[loc].append([laststart[loc], lastend[loc]])
```

```python
for loc, segs in timesegs.items():
    print(loc, ':', len(segs))
print(len(timesegs))
```

```python
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

plt.figure(figsize=(14, 10))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m/%d/%Y'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())

labels = []
for ind, (loc, segs) in enumerate(timesegs.items()):
    for seg in segs:
        seg_dates = [datetime.fromtimestamp(tt).date() for tt in seg]
        plt.plot(seg_dates, [ind, ind], color='C%d' % (ind % 10))
    labels.append(loc)
plt.yticks(list(range(len(labels))), labels)
plt.gcf().autofmt_xdate()
plt.show()
plt.close()
```
